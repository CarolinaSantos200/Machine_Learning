```python
from sklearn.linear_model import LogisticRegression
import pickle
from sklearn.metrics import accuracy_score, classification_report
from yellowbrick.classifier import ConfusionMatrix
```

```python
with open("credit.pkl","rb") as f:
    X_credit_training, y_credit_training, X_credit_test, y_credit_test = pickle.load(f)
```

```python
Logistic_Regression = LogisticRegression(random_state=1)
Logistic_Regression.fit(X_credit_training, y_credit_training)
```

```python
Logistic_Regression.intercept_
```

```python
Logistic_Regression.coef_
```

```python
predict = Logistic_Regression.predict(X_credit_test)
predict
```

```python
y_credit_test
```

```python
accuracy_score(y_credit_test,predict)
```

```python
cm = ConfusionMatrix(Logistic_Regression)
cm.fit(X_credit_training, y_credit_training)
cm.score(X_credit_test, y_credit_test)
```

```python
print(classification_report(y_credit_test, predict))
```
