```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from yellowbrick.classifier import ConfusionMatrix
import pickle 
```

```python
with open('credit.pkl', 'rb') as f:
  X_credit_training, y_credit_training, X_credit_test, y_credit_test = pickle.load(f)
```

```python
random_forest = RandomForestClassifier(n_estimators=40, criterion='entropy', random_state = 0)
random_forest.fit(X_credit_training, y_credit_training)
```

```python
predict = random_forest.predict(X_credit_test)
predict
```

```python
y_credit_test
```

```python
accuracy_score(y_credit_test, predict)
```

```python
cm = ConfusionMatrix(random_forest)
cm.fit(X_credit_training, y_credit_training)
cm.score(X_credit_test, y_credit_test)
```

```python
print(classification_report(y_credit_test, predict))
```
